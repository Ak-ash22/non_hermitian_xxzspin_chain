# D = 1

```python
hx = [1]*10
hz = np.linspace(-0.99*hx[0],hx[0],10,endpoint=False)
d = [1]*10
hzhx = []
for i in range(len(hx)):
    hzhx.append(hz[i]/hx[i])
```

### n = 3

```python
H3 = Hamiltonian(3,2,hx,hz,d)
S3 = biorthogonal_entropy(H3,1,1)
```

```python
plt.plot(hzhx,S3,color='r')
plt.show()
```
