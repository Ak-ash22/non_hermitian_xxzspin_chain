# D = 1

```python
hx = [1]*10
hz = np.linspace(-0.99*hx[0],hx[0],10,endpoint=False)
d1 = [1]*10
d2 = [-1]*10
hzhx = []
for i in range(len(hx)):
    hzhx.append(hz[i]/hx[i])
```

### n = 4

```python
H4 = Hamiltonian(4,2,hx,hz,d1)
H_4 = Hamiltonian(4,2,hx,hz,d2)
S4 = biorthogonal_entropy(H4,1,2)
S_4 = biorthogonal_entropy(H_4,1,2)
```

### n = 6

```python
H6 = Hamiltonian(6,3,hx,hz,d1)
H_6 = Hamiltonian(6,3,hx,hz,d2)
S6 = biorthogonal_entropy(H6,2,3)
S_6 = biorthogonal_entropy(H_6,2,3)
```

### n = 8

```python
H8 = Hamiltonian(8,4,hx,hz,d1)
H_8 = Hamiltonian(8,4,hx,hz,d2)
S8 = biorthogonal_entropy(H8,3,4)
S_8 = biorthogonal_entropy(H_8,3,4)
```

```python
H10 = Hamiltonian(10,5,hx,hz,d1)
S10 = biorthogonal_entropy(H10,4,5)
```
