## D = -1

```python
hx1 = [4]*100
hz1 = np.linspace(-0.1*hx1[0],0.1*hx1[0],100,endpoint=False)       
d1 = [-1]*100
hzhx1 = []
for i in range(len(hx1)):
    hzhx1.append(hz1[i]/hx1[i])
```

```python
H1 = Hamiltonian(3,2,hx1,hz1,d1)
S1 = biorthogonal_entropy(H1,1,1)
```
