```python
a = [-1,-0.5,0,1,10]
for k in range(len(a)):
    hz = np.linspace(-4,4,1000,True)
    hx =[2]*1000
    D = [a[k]]*1000
    hzhx = []
    #hxD = []
    hx_plot = []
    hz_plot = []
    lmbda_real = []
    lmbda_im = []
    coords = [hzhx,lmbda_im]
    for i in range(len(hx)):
        j = 8
        while j!=0:
            #Dhx.append(D[i]/hx[i])
            #hxD.append(hx[i]*D[i])
            hx_plot.append(hx[i])
            hz_plot.append(hz[i])
            hzhx.append(hz[i]/hx[i])
            j-=1  
    #print(np.shape(hzhx))
        H_c = 1j*hz[i]*sz2 + hx[i]*sx2
        H_ab = sx_ab + sy_ab + D[i]*sz_ab
        H_bc = sx_bc + sy_bc + D[i]*sz_bc
        H_ac = sx_ac + sy_ac + D[i]*sz_ac
        H_t = H_c + H_ab + H_bc + H_ac
        lmbda = np.linalg.eigvals(H_t)
        for k in range(len(lmbda)):
            lmbda_real.append(lmbda.real[k])
            lmbda_im.append(np.round(lmbda.imag[k],5))
            np.shape(lmbda.real)
    #print(np.round(hzD,4))
    plt.scatter(hzhx,lmbda_im,s=5)
    plt.ylabel('imaginary roots')
    plt.xlabel('hz/hx')
    plt.title('D = {}'.format(D[i]))
    plt.show()
```
