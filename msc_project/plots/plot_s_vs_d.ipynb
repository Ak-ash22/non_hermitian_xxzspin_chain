```python
hx = [1]*100
hz1 = np.linspace(-hx[0],hx[0],num=100,endpoint=False)    #Case 1: hz<hx
hz2 = np.linspace(hx[0]-0.01,hx[0],num=100,endpoint=False) #Case 2: hz~hx
D = np.linspace(-1,1,num=100,endpoint=True)
```

```python
H1 = H(hx,hz1,D)
H2 = H(hx,hz2,D)
```

```python
print(np.round(np.linalg.eigvals(H1).imag,4))
```

```python
for i in range(len(H2)):
    print(np.round(np.linalg.eigvals(H2[i]).imag,4))
```

```python
S_rc1, S_lc1 = S_c(H1)
S_rc2, S_lc2 = S_c(H2)
```

```python
np.round(S_lc2,10).all() == np.round(S_rc2,10).all()
```

```python
plt.subplot(2, 1, 1) # row 1, col 2 index 1
plt.plot(D, S_rc1, color = 'r')
plt.title('Case 1: hz<hx')
plt.xlabel('Delta (D)')
plt.ylabel('Entanglement entropy (S_c)')

plt.subplot(2, 1, 2) # index 2
plt.plot(D, S_rc2, color = 'g')
plt.title("Case 2: hz~hx")
plt.xlabel('Delta (D)')
plt.ylabel('Entanglement entropy (S_c)')

plt.subplots_adjust(hspace=0.5,wspace=2)
plt.show()
```

```python
plt.plot(D, S_rc1, color = 'r', label = 'hz<hx')
plt.plot(D, S_rc2, color = 'g', label = 'hz~hx')
plt.xlabel('Delta')
plt.ylabel('Entanglement entropy S_rc')
plt.legend()
plt.show()
```

```python
plt.subplot(2, 1, 1) # row 1, col 2 index 1
plt.plot(D, S_lc1, color = 'r')
plt.title('Case 1: hz<hx')
plt.xlabel('Delta (D)')
plt.ylabel('Entanglement entropy (S_c)')

plt.subplot(2, 1, 2) # index 2
plt.plot(D, S_lc2, color = 'g')
plt.title("Case 2: hz~hx")
plt.xlabel('Delta (D)')
plt.ylabel('Entanglement entropy (S_c)')

plt.subplots_adjust(hspace=0.5,wspace=2)
plt.show()
```
