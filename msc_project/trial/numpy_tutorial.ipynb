#### Integrals and derivatives

```python
x = np.linspace(1,10,100)
y = 1/x**2 + np.sin(x)
```

```python
plt.plot(x,y)
```

```python
dydx = np.gradient(y,x)
plt.plot(x,y)
plt.plot(dydx)
```

```python
#Discrete Integration
y_int = np.cumsum(y)*(x[1]-x[0]) # integration is nothing but cumulative sum of y's multipled by delta x.
plt.plot(x,y)
plt.plot(dydx)
plt.plot(y_int)
```
